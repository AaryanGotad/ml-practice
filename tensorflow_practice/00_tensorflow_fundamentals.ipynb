## Creating Random Tensors

## Exercise

There are 4 ways random operations are performed.

1. Neither the global nor the operation seed is given
2. Global seed is set but operation seed is not
3. Operation seed is set but global seed is not
4. Lastly, both glovbal and operation seed are set

#### First, lets look what will happen if neither global seed nor the operation seed is set

In this case we get different results for every call to the random op **AND** every re-run of the program.

```python
print(tf.random.uniform([1]))
print(tf.random.uniform([1]))
```

_(Now closing the program and re running it)_

```python
print(tf.random.uniform([1]))
print(tf.random.uniform([1]))
```

### Second, if the global seed is set but the operation seed is not

Here we get different results for every call to the random op, but the sequence would stay the same for every re-run of the program

```python
tf.random.set_seed(1234)
print(tf.random.uniform([1]))
print(tf.random.uniform([1]))
```

_(Now closing the program and re running it)_

```python
tf.random.set_seed(1234)
print(tf.random.uniform([1]))
print(tf.random.uniform([1]))
```

_The reason we get_ "[0.3253647]" _instead of_ "[0.5380393]" _on the second call of `tf.random.uniform` above is because the second call uses a different operation seed._

### Third, if the operation seed is set but the global seed is not

We get different results for every call to the random op, but the same sequence for every re-run of the program.

_(Similiar to the second case)_

```python
print(tf.random.uniform([1], seed=1))
print(tf.random.uniform([1], seed=1))
```

_(Now closing the program and re-running it)_

```python
print(tf.random.uniform([1], seed=1234))
print(tf.random.uniform([1], seed=1234))
```

### Fourth, the last case when both the global and operation seed are set

Here are two cases,
* The global seed and operation seed are different-
  * Here the results are the same as the third and second case
* The global seed and operation seed are same-
  * Here's where things change, individual random op this provide random results, but the sequence is same, even if we close the program and run it again

```python
tf.random.set_seed(1234)
print(tf.random.uniform([1], seed=1234))
print(tf.random.uniform([1], seed=1234))
```

```python
tf.random.set_seed(1234)
print(tf.random.uniform([1], seed=1234))
print(tf.random.uniform([1], seed=1234))
```

Now using all the knowledge we gained about random operations and how global and operation seeds work, lets shuffle some tensors with it shall we!

```python
tf.random.set_seed(12)
tf.random.shuffle(not_shuffled, seed=12)
```

In this case the order of shuffling stays same as well as the randomness of the individual random op, if there is a difference in either of the two seeds then only the order would be the same but not the individual random operation

### Getting information from tensors

When dealing with tensors we probably want to be aware of the following attributes:
* Shape: The length (number of elements) of each of the dimensions of a tensor.
  * Code: `tensor.shape`
* Rank: The number of tensor dimensions. A scalar has rank 0, a vector has rank 1, a matrix is rank 2, a tensor has rank n.
  * Code: `tensor.ndim`
* Axis or dimension: A particular dimension of a tensor.
  * Code: `tensor[0]`, `tensor[:, 1]`
* Size: The total number of items in the tensor.
  * Code: `tf.size(tensor)`

```python
# Create a rank 4 tensor (4 dimensions)
rank_4_tensor = tf.zeros(shape=[2, 3, 4, 5])
rank_4_tensor
```

```python
rank_4_tensor[0]
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
2 * 3 * 4 * 5
```

```python
# Get various attributes of our tensor
print("Datatype of every elements:", rank_4_tensor.dtype)
print("Number of dimensions (rank):", rank_4_tensor.ndim)
print("Shape of tensor:", rank_4_tensor.shape)
print("Elements along zero axis:", rank_4_tensor.shape[0])
print("Elements along the last axis:", rank_4_tensor.shape[-1])
print("Total number of elements in our tensor:", tf.size(rank_4_tensor))
print("Total number of elements in our tensor:", tf.size(rank_4_tensor).numpy())
```

**The Dot Product**

Matrix multiplication is also referred to as the dot product
You can perform matrix multiplication using:
* `tf.matmul()`
* `tf.tensordot()`
* `@`

```python
X = tf.constant([[1, 2],
                 [3, 4],
                 [5, 6]])
Y = tf.constant([[7, 8],
                 [9, 10],
                 [11, 12]])

X, Y
```

```python
# Perform the dot product on X and Y (transposed)
tf.tensordot(tf.transpose(X), Y, axes=1)
```

```python
tf.tensordot(tf.reshape(X, shape=(2, 3)), Y, axes=1)
```

```python
# Perform matrix multiplication between X and Y (reshaped)
tf.matmul(X, tf.reshape(Y, shape=(2, 3)))
```

```python
tf.matmul(X, tf.transpose(Y))
```

```python
# Check the values of Y, reshape Y and transposed Y
print("Normal Y:")
print(Y, "\n")

print("Y reshaped to (2, 3):")
print(tf.reshape(Y, (2, 3)), "\n")

print("Y transpose:")
print(tf.transpose(Y), "\n")
```

Generally, when performing matrix multiplication and one of the axes doesn't line up, you will transpose rather than reshape one of the tensors to satisfy the matrix multiplication rules.

**Find the positional maximum and minimum of a tensor**

Positional maximum/minimum means at what position we get the maximum/minimum value respectively of the tensor

```python
# Creating a new tensor for finding positional maximum and minimum
F = tf.random.uniform(shape=[50])
F
```

```python
# Finding the positional maximum
tf.argmax(F)
```

```python
# Indexing on our largest value position
F[tf.argmax(F)]
```

```python
# Finding the max value of F
tf.reduce_max(F)
```

```python
# Checking for equality (Just to be sure)
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# Finding the positional minimum
tf.argmin(F)
```

```python
F[tf.argmin(F)]
```

```python
# Finding the min value of our tensor
tf.reduce_min(F)
```

```python
# Checking for equality
F[tf.argmin(F)] == tf.reduce_min(F)
```
